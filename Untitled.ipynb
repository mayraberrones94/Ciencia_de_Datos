```python
import statsmodels.api as sm
from numpy import isnan
import pandas as pd
import ssl

if getattr(ssl, '_create_unverified_context', None):
    ssl._create_default_https_context = ssl._create_unverified_context
data = pd.read_csv("https://raw.githubusercontent.com/mayraberrones94/Ciencia_de_Datos/master/bp_w-data.csv")

# Quitar variables que no me sirven
data = data.drop(['Unnamed: 32','id'],axis = 1)

# Para poder usar el B y M como parámetro objetivo
data.diagnosis.replace(to_replace = dict(M = 1, B = 0), inplace = True)
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
from numpy import isnan
import pandas as pd
import ssl

mal = data.loc[data.diagnosis == 1]
ben = data.loc[data.diagnosis == 0]

f = plt.figure()  
d = mal
d = pd.DataFrame(d, columns = ["radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean", "compactness_mean", "concavity_mean", "concave points_mean", "symmetry_mean", "fractal_dimension_mean"])
d = d.dropna() 
print(d.corr())
sf = f.add_subplot(211)
tmp = sf.matshow(d.corr())

d = ben
d = pd.DataFrame(d, columns = ["radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean", "compactness_mean", "concavity_mean", "concave points_mean", "symmetry_mean", "fractal_dimension_mean"])
d = d.dropna() 
print(d.corr())
sf = f.add_subplot(212)
tmp = sf.matshow(d.corr())
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
from numpy import isnan
import pandas as pd
import ssl

mal = data.loc[data.diagnosis == 1]
ben = data.loc[data.diagnosis == 0]

f = plt.figure()  
d = mal
d = pd.DataFrame(d, columns = ["radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean"])
d = d.dropna() 
print(d.corr())
sf = f.add_subplot(211)
tmp = sf.matshow(d.corr())

d = ben
d = pd.DataFrame(d, columns = ["radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean"])
print(d.corr())
sf = f.add_subplot(212)
tmp = sf.matshow(d.corr())
```

```python
mal = data.loc[data.diagnosis == 1]
d = mal
d = pd.DataFrame(d, columns = ["radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean"])
print(len(d))
tmp = pd.plotting.scatter_matrix(d, figsize = (9, 9), s = 500) 
y = d["radius_mean"]
x = d[["texture_mean", "perimeter_mean", "area_mean", "smoothness_mean"]]
x = sm.add_constant(x) 
lol = sm.OLS(y, x).fit()  
print(lol.summary())
```
